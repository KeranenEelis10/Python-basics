```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('Acc teht1.csv',delimiter=";", decimal=',')
```

```python
df = df.rename(columns={"Time (s)":"time","Linear Acceleration x (m/s^2)":"ax", "Linear Acceleration y (m/s^2)":"ay","Linear Acceleration z (m/s^2)":"az","Absolute acceleration (m/s^2)":"a_tot" })
df.info()
df.head()
```

```python
fig = plt.figure(num=None, figsize= (20, 10),dpi=100, facecolor='w', edgecolor='k')

plt.subplot(3,1,1)
plt.plot(df.time,df.ax)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Linear acceleration x')
plt.axis([0,30,-55,55])

plt.subplot(3,1,2)
plt.plot(df.time,df.ay)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Linear acceleration x')
plt.axis([0,30,-50,50])

plt.subplot(3,1,3)
plt.plot(df.time,df.az)
plt.grid()
plt.xlabel('Time')
plt.ylabel('Linear acceleration x')
plt.axis([0,30,-50,50])

plt.show()
```
